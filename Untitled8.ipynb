Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.

Answer- Difference between t-test and z-test:

t-test: It is used when the sample size is small (typically less than 30) or when the population standard deviation is unknown. It assesses whether the means of two groups are statistically different from each other.
z-test: It is used when the sample size is large (typically greater than 30) and when the population standard deviation is known. It compares a sample mean to a known population mean.
Example scenario:

t-test: You might use a t-test to compare the average exam scores of students who received a new teaching method versus those who received the traditional teaching method.
z-test: You might use a z-test to compare the average income of employees in your company with the average income of employees nationally.

Q2: Differentiate between one-tailed and two-tailed tests.

Answer- One-tailed vs. two-tailed tests:

One-tailed test: It tests for a difference in one direction only (e.g., greater than or less than). It's appropriate when you're interested in knowing if a parameter is either significantly greater or significantly less than a certain value.
Two-tailed test: It tests for a difference in both directions (e.g., not equal to). It's appropriate when you're interested in knowing if a parameter is significantly different from a certain value, but you don't have a specific direction in mind.

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

Answer-Type 1 and Type 2 errors:

Type 1 error: It occurs when you reject a true null hypothesis (false positive).
Type 2 error: It occurs when you fail to reject a false null hypothesis (false negative).
Example scenario:

Type 1 error: Convicting an innocent person in a court trial.
Type 2 error: Declining a potentially effective medical treatment.

Q4: Explain Bayes's theorem with an example.

Answer- Bayes's theorem:
Bayes's theorem describes the probability of an event, based on prior knowledge of conditions that might be related to the event.

Example:
Suppose there's a disease that affects 1% of the population. A test for this disease correctly identifies it 99% of the time when the disease is present, but it also has a 5% false positive rate. You can use Bayes's theorem to calculate the probability of having the disease given a positive test result.

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

Answer- Confidence interval:
A confidence interval is a range of values that likely contains the true population parameter. It's calculated from the sample data and provides a range within which the true population parameter is likely to lie with a certain level of confidence.

Example:
Let's say you have a sample of 100 students, and you want to estimate the average height of all students in the school with 95% confidence. If the sample mean height is 65 inches with a standard deviation of 3 inches, you can calculate the confidence interval for the population mean height.

Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

Answer-Margin of error in a confidence interval:
The margin of error is the range above and below the sample statistic within which the true population parameter is estimated to lie with a certain level of confidence. It decreases as the sample size increases because larger samples provide more precise estimates of the population parameter.

Example:
If you're conducting a political survey with a sample size of 1000 people and you find that 60% of them support a particular policy with a margin of error of 3%, this means that you're 95% confident that the true proportion of people who support the policy lies between 57% and 63%.
